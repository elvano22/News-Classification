# Crawling dan Scraping

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import datetime
```

## Crawling kompas.com

```python
links = []

indexes = ['travel','bola','nasional', 'umkm']

for i in range (1,300):
  if len(links) > 100: break;

  for index in indexes:
    if len(links) > 100: break;

  # URL dari Kompas.com
  url = f'https://indeks.kompas.com/?site={index}&page={i}'

  # Mengambil halaman dengan requests
  response = requests.get(url)

  # Parsing konten halaman menggunakan BeautifulSoup
  soup = BeautifulSoup(response.text, 'html.parser')

  # Mencari semua elemen berita utama, misalkan menggunakan class yang spesifik pada elemen berita
  articles = soup.find_all('div', class_='articleItem')

  # Mencari link dari setiap berita
  for article in articles:
    link = article.find('a', class_='article-link')['href']
    links.append(link)

# Melihat 5 link pertama
links[:5]
```

```python
len(links)
```

## Scraping kompas.com

```python
labels =[]
bodies = []
medias = []
for link in links:
  # Mengambil halaman dengan requests
  response = requests.get(f'{link}?page=all#page')
  if response.status_code != 200: continue

  # Parsing konten halaman menggunakan BeautifulSoup
  soup = BeautifulSoup(response.text, 'html.parser')

  # Mencari semua isi berita
  body_temp = soup.find_all('p')
  body_temp
  body_texts=[]
  for x in body_temp:
    if not x.has_attr('class'):
      body_texts.append(x.get_text().replace("\xa0", " "))

  # Mencari label
  labels_temp = soup.find_all('li', class_='breadcrumb__item')

  if len(labels_temp) == 0 or len(body_texts)<2: continue
  labels.append(labels_temp[len(labels_temp)-1].get_text())
  bodies.append(" ".join(body_texts))
  medias.append("Kompas")
```

```python
# Membuat dataframe
df = pd.DataFrame({
    'Label': labels,
    'Media': medias,
    'Body': bodies
})
df
```

## Crawling liputan6.com

```python
links_2=[]

# Tanggal awal yang ingin Anda mulai dari tanggal terbaru
start_date = datetime.date.today()

indexes = ['showbiz', 'bola', 'bisnis', 'pemilu']

for i in range (1,300):
  if len(links_2) > 100: break;

  for index in indexes:
    if len(links_2) > 100: break;

    # URL dari liputan6.com
    current_date = start_date - datetime.timedelta(days=i)
    url = f'https://www.liputan6.com/{index}/indeks/{current_date.year}/{current_date.month:02d}/{current_date.day:02d}'

    # Mengambil halaman dengan requests
    response = requests.get(url)

    # Parsing konten halaman menggunakan BeautifulSoup
    soup = BeautifulSoup(response.text, 'html.parser')

    # Mencari semua elemen berita utama, misalkan menggunakan class yang spesifik pada elemen berita
    articles = soup.find_all('article', class_='articles--rows--item')

    # Mencari link dari setiap berita
    for article in articles:
      link = article.find('a', class_='articles--rows--item__title-link')['href']
      links_2.append(link)

# Melihat 5 link pertama
links_2[:5]
```

## Scraping liputan6.com

```python
labels =[]
bodies = []
medias = []
for link in links_2:
  # Mengambil halaman dengan requests
  response = requests.get(link)
  if response.status_code != 200: continue

  # Parsing konten halaman menggunakan BeautifulSoup
  soup = BeautifulSoup(response.text, 'html.parser')

  # Mencari semua isi berita
  body_temp = soup.find_all('p')
  body_temp
  body_texts=[]
  for x in body_temp:
    if not x.has_attr('class'):
      body_texts.append(x.get_text().replace("\xa0", " "))

  # Mencari label
  labels_temp = soup.find_all('li', class_='read-page--breadcrumb--item')

  if len(labels_temp) == 0 or len(body_texts)<2: continue
  labels.append(labels_temp[len(labels_temp)-1].get_text())
  bodies.append(" ".join(body_texts))
  medias.append("Liputan6")
```

## Crawling detik.com

```python
links_3=[]

indexes = ['https://sport.detik.com/', 'https://travel.detik.com/', 'https://finance.detik.com/ekonomi-bisnis', 'https://news.detik.com/pemilu']

for i in range (1,300):
  if len(links_3) > 100: break;

  for index in indexes:
    if len(links_3) > 100: break;

  # URL dari detik.com
  url = f'{index}/indeks/{i}'

  # Mengambil halaman dengan requests
  response = requests.get(url, allow_redirects=False)

  # Parsing konten halaman menggunakan BeautifulSoup
  soup = BeautifulSoup(response.text, 'html.parser')

  # Mencari semua elemen berita utama, misalkan menggunakan class yang spesifik pada elemen berita
  articles = soup.find_all('article', class_='list-content__item')

  # Mencari link dari setiap berita
  for article in articles:
    link = article.find('a', class_='media__link')['href']
    links_3.append(link)

# Melihat 5 link pertama
links_3[:5]
```

```python
len(links_3)
```

## Scraping detik.com

```python
labels =[]
bodies = []
medias = []
for link in links_3:
  # Mengambil halaman dengan requests
  response = requests.get(link)
  if response.status_code != 200: continue

  # Parsing konten halaman menggunakan BeautifulSoup
  soup = BeautifulSoup(response.text, 'html.parser')

  # Mencari semua isi berita
  body_temp = soup.find_all('p')
  body_temp
  body_texts=[]
  for x in body_temp:
    if not x.has_attr('class'):
      body_texts.append(x.get_text().replace("\xa0", " "))

  # Mencari label
  labels_temp = soup.select('div.page__breadcrumb a')

  # Menambah media
  if len(labels_temp) == 0 or len(body_texts)<2: continue
  labels.append(labels_temp[len(labels_temp)-1].get_text().strip())
  bodies.append(" ".join(body_texts))
  medias.append("Detik")
```

```python
# Membuat dataframe
df_detik = pd.DataFrame({
    'Label': labels,
    'Media': medias,
    'Body': bodies
})
df_detik
```

```python
# Definisi daftar untuk masing-masing kategori
politik = ['politik', 'nasional', 'pemilu', 'pemerintah', 'pemilu']
olahraga = ['timnas indonesia', 'liga italia', 'bulutangkis', 'olahraga', 'liga inggris', 'liga indonesia', 'badminton', 'sports', 'sport', 'bola', 'liga italia', 'liga internasional']
bisnis = ['e-business', 'swasta', 'bisnis', 'spend smart', 'earn smart', 'smartpreneur', 'housing', 'ekonomi', 'program', 'jagoan lokal', 'training', 'ekonomi', 'bank']
hiburan = ['hype', 'whats new', 'stori', 'feel good', 'hiburan', 'travel', 'travel update', 'aksesoris', 'jalan jalan', 'travel update', 'tempat makan', 'resep', 'game', 'tips kuliner', 'musik', 'showbiz', 'selebritis', 'blink-blink', 'k-pop']

# Fungsi untuk menentukan kategori berdasarkan label
def determine_category(label):
  if label is None:
    return None
  label = label.lower()
  if label in politik:
    return 'politik'
  elif label in olahraga:
    return 'olahraga'
  elif label in bisnis:
    return 'bisnis'
  elif label in hiburan:
    return 'hiburan'
  else:
    return None

# Terapkan fungsi pada kolom label
df_merged['Label'] = df_merged['Label'].apply(determine_category)

# Membuang baris yang kategorinya menjadi None setelah penggantian
df_merged.dropna(subset=['Label'], inplace=True)
df_merged.reset_index(drop=True, inplace=True)

# Tampilkan hasil DataFrame yang telah diperbarui
df_merged
```

## Cleansing

```python
def clean_text(text):
  # Menghapus kata "Advertisement" (case insensitive)
  text = re.sub(r'\bAdvertisement\b', '', text)
  # Menghapus kalimat terakhir dari "* Fakta atau Hoaks?" sampai akhir
  text = re.sub(r'\* Fakta atau Hoaks\?[\s\S]*$', '', text)
  # Mengubah semua teks menjadi lowercase
  text = text.lower()
  # Membuat pattern regex untuk menemukan "KOMPAS.com - " atau "KOMPAS.com — " tanpa memperdulikan case
  pattern = re.compile(r'kompas\.com\s*[-—]\s*')
  # Menghapus teks sebelum pattern
  cleaned = pattern.split(text, 1)[1] if pattern.search(text) else text
  # Menghapus bagian akhir mulai dari " Copyright"
  cleaned = re.split(r'\s*copyright', cleaned, 1)[0]
  # Menghilangkan semua tanda baca
  cleaned = re.sub(r'[^\w\s]', '', cleaned)
  # Menghapus semua angka
  cleaned = re.sub(r'\d+', '', cleaned)
  return cleaned

# Menerapkan fungsi ke kolom teks
df_merged['Cleaned Text'] = df_merged['Body'].apply(clean_text)
df_merged
```

## Tokenization dan menghapus stop words

```python
# Inisialisasi stop words remover untuk bahasa Indonesia
stop_factory = StopWordRemoverFactory()
stopwords = stop_factory.get_stop_words()

def tokenize_text(text):
  tokens = word_tokenize(text)
  filtered_tokens = [word for word in tokens if word.lower() not in stopwords]
  return filtered_tokens

df_merged['Cleaned Text'] = df_merged['Cleaned Text'].apply(tokenize_text)
df_merged
```

## Stemming

```python
stemmer = StemmerFactory().create_stemmer()

def stem_tokens(tokens):
  stemmed_tokens = [stemmer.stem(token) for token in tokens]
  return stemmed_tokens

df_merged['Cleaned Text'] = df_merged['Cleaned Text'].apply(stem_tokens)
df_merged
```

# Text Representation

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from gensim.models import Word2Vec
import numpy as np
```

## Metode 1: TF-IDF

TF-IDF, which stands for Term Frequency-Inverse Document Frequency, is a method for calculating the weight of a word in a document that reflects the importance of that word in that document relative to the data set.

```python
# Menggabungkan token yang telah distem ke dalam kalimat kembali
df_merged['Processed Text'] = df_merged['Cleaned Text'].apply(lambda x: ' '.join(x))

# Inisialisasi TF-IDF Vectorizer
tfidf_vectorizer = TfidfVectorizer()
tfidf_matrix = tfidf_vectorizer.fit_transform(df_merged['Processed Text'])
```

```python
# Mendapatkan nama fitur/kata dari TF-IDF Vectorizer
feature_names = tfidf_vectorizer.get_feature_names_out()

# Membuat DataFrame dari matriks TF-IDF
tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=feature_names)
tfidf_df
```

## Metode 2: Word2Vec

Word2Vec is a model that trains word vector representations from a collection of text. The model generates word vectors that reflect the semantic context of words; words that occur frequently in the same context have similar vector representations. Two architectures in word2vec:

- CBOW (Continuous Bag of Words)

  This model predicts the target word based on the surrounding words (context).

- Skip-Gram

  The opposite of CBOW, this model uses a single target word to predict surrounding words.

```python
sentences = df_merged['Cleaned Text'].tolist()

# Initialize and train the Word2Vec model
model_w2v = Word2Vec(sentences, vector_size=50, window=5, min_count=3, sg=1)

# Save and use the model
word_vectors = model_w2v.wv

i=0
for word in word_vectors.key_to_index:
  if(i>5): break
  vector = word_vectors[word]
  print(f"Word: {word} \nVector: {vector}\n")
  i+=1
```

```python
def document_vector(doc):
  # Menghapus kata yang tidak ada di model
  words = [word for word in doc if word in model_w2v.wv.key_to_index]
  if len(words) >= 1:
    return np.mean(model_w2v.wv[words], axis=0)
  else:
    return []

# Menghitung vektor rata-rata untuk setiap dokumen
df_merged['doc_vector'] = df_merged['Cleaned Text'].apply(document_vector)

# Menghapus baris yang tidak bisa di vektorisasi (jika ada)
df_merged = df_merged[df_merged['doc_vector'].map(len) > 0]
X_w2v = np.array(list(df_merged['doc_vector']))
X_w2v
```

# Classification Model

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import classification_report
```

## Train Test Split

```python
X_train_tfidf, X_test_tfidf, y_train, y_test = train_test_split(tfidf_matrix, df_merged['Label'], test_size=0.2, random_state=99)
X_train_w2v, X_test_w2v = train_test_split(X_w2v, test_size=0.2, random_state=99)
```

```python
len(X_w2v)
```

```python
# Mendefinisikan parameter untuk tuning
param_svm = {'C': [0.1, 1, 10], 'gamma': [0.001, 0.01, 0.1, 1]}
param_rf = {'n_estimators': [50, 100, 200], 'max_depth': [10, 20, 30]}
```

## SVM with TF-IDF

```python
svm_tfidf = GridSearchCV(SVC(), param_svm, cv=5)
svm_tfidf.fit(X_train_tfidf, y_train)
svm_tfidf_pred = svm_tfidf.predict(X_test_tfidf)

# Evaluasi model
svm_tfidf_report = classification_report(y_test, svm_tfidf_pred,  digits=5, output_dict=True)
print("SVM dengan TF-IDF:")
print(classification_report(y_test, svm_tfidf_pred,  digits=5))
```

## Random Forest with TF-IDF

```python
rf_tfidf = GridSearchCV(RandomForestClassifier(), param_rf, cv=5)
rf_tfidf.fit(X_train_tfidf, y_train)
rf_tfidf_pred = rf_tfidf.predict(X_test_tfidf)

rf_tfidf_report = classification_report(y_test, rf_tfidf_pred, digits = 5, output_dict=True)
print("Random Forest dengan TF-IDF:")
print(classification_report(y_test, rf_tfidf_pred, digits = 5))
```

## SVM with Word2Vec

```python
svm_w2v = GridSearchCV(SVC(), param_svm, cv=5)
svm_w2v.fit(X_train_w2v, y_train)
svm_w2v_pred = svm_w2v.predict(X_test_w2v)

svm_w2v_report = classification_report(y_test, svm_w2v_pred, digits=5, output_dict=True)
print("SVM dengan Word2Vec:")
print(classification_report(y_test, svm_w2v_pred, digits=5))
```

## Random Forest with Word2Vec

```python
rf_w2v = GridSearchCV(RandomForestClassifier(), param_rf, cv=5)
rf_w2v.fit(X_train_w2v, y_train)
rf_w2v_pred = rf_w2v.predict(X_test_w2v)

rf_w2v_report = classification_report(y_test, rf_w2v_pred, digits=5, output_dict=True)
print("Random Forest dengan Word2Vec:")
print(classification_report(y_test, rf_w2v_pred, digits=5))
```
